```python
#Exercise 1
# Custom function that takes Price as input return $ if its value is less than 10 $$ less than 30 and $$$ greater than 30
def yelp(price):
    if price >10 and price <20:
        return "$"
    elif price >20 and price <30:
        return "$$"
    elif price > 30:
        return "$$$"
```

```python
# Write a custom function yelp to apply on the total_bill column and display $ if dollarprice <10, $$$ <30 , $$$ >30
dollarprice = df["total_bill"].apply(yelp)
df["dollarprice"] = dollarprice
df
```

```python
# defining the custom function and classifying the data category.
def quality(total_bill1,tip1):
    if tip1 > total_bill1*0.25:
        return "Generous"
    else:
        return "Others"
    
```

```python
quality(100,30)
```

```python
# Write a custom function taking multiple columns total_bill and tip  as input 
# and caluclate the tip % 25 if >= 25 return Generous else return Others
df["Quality"]=df[["total_bill","tip"]].apply(lambda df: quality(df["total_bill"],df["tip"]),axis=1)
```

```python
df
```

```python
# Remember performance
# Caluclate the quality function using vectorize by taking 2 columns as input total_bill and tip

df["Quality"]= np.vectorize(quality)(df["total_bill"], df["tip"])
# this works way faster &  its more better convenicence when compare to the lambda function.
# np.vectorize(custom function)(input1, input2)
# np.vectorise tranforms into numpy arrays, as we know numpys are faster



```

```python
setup ='''
import numpy as np
import pandas as pd
df =pd.read_csv('./course2/DATA/tips.csv')
def quality(total_bill1,tip1):
    if tip1 > total_bill1*0.25:
        return "Generous"
    else:
        return "Others"
'''
```

```python
stmt1= '''df["Quality"]=df[["total_bill","tip"]].apply(lambda df: quality(df["total_bill"],df["tip"]),axis=1)'''
stmt2= '''df["Quality"]= np.vectorize(quality)(df["total_bill"], df["tip"])'''
```

```python
timeit.timeit(setup=setup,stmt= stmt1 ,number=100)
```
