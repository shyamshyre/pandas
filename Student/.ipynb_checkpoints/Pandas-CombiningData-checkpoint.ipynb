## Concatenation of data

```python
# Take a data set dictionary with 'A,B' as key.
# list of 'A0 ', 'A1', 'A2', 'A3' as values 
# list of 'B0', 'B1', 'B2', 'B3' as values 


data_one = {'A': ['A0', 'A1', 'A2', 'A3'],'B': ['B0', 'B1', 'B2', 'B3']}
```

```python
# Take a data set dictionary with 'C,D' as key.
# list of 'C0 ', 'C1', 'C2', 'C3' as values 
# list of 'D0', 'D1', 'D2', 'D3' as values 

data_two = {'C': ['C0', 'C1', 'C2', 'C3'], 'D': ['D0', 'D1', 'D2', 'D3']}
```

```python
# Build a data frame with one ,two with the above two data sets.

one = pd.DataFrame(data_one)
two = pd.DataFrame(data_two)
```

# Merging the data  (Data Tables)

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

# pd.merge()

##### Merge pandas DataFrames based on key columns, similar to a SQL join. Results based on the how parameter.

```python
help(pd.merge)
```

```python
# Merge takes the two data frames data and peform what type of join which columns

pd.merge(registrations,logins,how="inner", on="name")
```

```python
# Display only common data across both the tables
# This is also returning the same data just columsn interchanged.
pd.merge(logins,registrations, how="inner", on="name")
```

```python
# Perform Left merge on registration,login tables on top of column "name"
# left merge: Returns the left data and the common data but not the right data
pd.merge(registrations,logins,how="left",on="name")
```

```python
# Perform Right merge on registration,login tables on top of column "name"
# Right merge : Display all the right side Dataframe elements and the common elments of both left and right DataFrame.
pd.merge(registrations,logins,how="right",on="name")
```

```python
# Perform Right merge on registration,login tables on top of column "name"
pd.merge(registrations,logins,how="right",on="name")
```

```python
# Display all the data across both the table, but dont display identical data twice, use outter join
# This will return all the data across both the tables but doesnt return commonality data twice
pd.merge(registrations,logins,how="outer",on="name")
```

```python
# Change the registration data frame index to name
# Now the index of the Table/Dataframe is changed to the name instead of id
registrations = registrations.set_index('name')
```

```python
registrations
```

```python
logins
```

```python
# index: Represents the index 
# on: Represents the column
# left_index:Left Table/DF index
# right_on:Right Table/DF column
# how -> type of relation

pd.merge(registrations,logins,left_index=True,right_on="name",how="inner")
```

```python
# Reset the index value as it is
registrations.reset_index()
```
